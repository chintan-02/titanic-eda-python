# TITANIC DATASET - EXPLORATORY DATA ANALYSIS

```python
## Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load the dataset

df = pd.read_csv('titanic.csv')
```

```python
## Show all column names

df.columns
```

```python
## Get summary statistics of the dataset

df.describe()
```

```python
## Check the data types of each column

df.info()
```

```python
## Check for missing values in each column

df.isnull().sum()
```

```python
## Visualize the distribution of passengers by gender

gender_counts = df['Sex'].value_counts()

plt.figure()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
## Visualize the distribution of passengers by class

sns.countplot(x='Survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
## Visualize the survival rate by gender

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
## Visualize the survival rate by passenger class

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
## Visualize the distribution of ages

sns.kdeplot(data=df['Age'].dropna(), fill=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
## Visualize the correlation matrix for numeric features

plt.figure(figsize=(10,6))  # Better size for visibility

corr = df.corr(numeric_only=True)
mask = np.triu(corr)  # Hide upper triangle

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm',
    fmt=".2f",              # Show 2 decimal points
    linewidths=0.5
)

plt.title("Correlation Matrix (Numeric Features)", fontsize=14)
plt.show()
```

```python
# Step 15: Key Observations from EDA
print("EDA Observations:")

print("1. Missing values are present in key features like Age, Cabin, and Embarked, which will require proper handling before model training.")

print("2. Survival is imbalanced, indicating a classification problem with unequal class distribution.")

print("3. Gender shows a strong relationship with survival, suggesting females had a higher survival rate compared to males.")

print("4. Passenger class (Pclass) is negatively correlated with survival, indicating higher-class passengers had better chances of survival.")

print("5. Age distribution shows most passengers were between young adults and middle age, with fewer elderly passengers.")

print("6. Fare appears positively correlated with survival, suggesting wealthier passengers had higher survival probability.")

print("7. The correlation heatmap indicates that no single feature strongly determines survival, meaning multiple features contribute to predictions.")
```
